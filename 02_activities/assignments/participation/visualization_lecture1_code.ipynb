```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
font1 = {'family': 'sans-serif', 'color': 'blue', 'size': '20'}
font2 = {'family': 'monospace', 'color': 'green', 'size': '40'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y) 
ax.set_title("line plot", fontdict= font1, loc='left')
ax.set_xlabel("x", fontdict= font2, loc='right')
ax.set_ylabel('y', fontdict= font2, loc='center')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,
           y,
           marker='*',
           color='indigo'
           )
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x ,y, marker='*', 
        color='red', 
        linestyle = '--', 
        linewidth = '2', 
        markeredgecolor = 'blue', 
        markerfacecolor = 'orange',
        markersize=12)

fig.show()

```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.grid(axis='y', color = 'black', linewidth=1, linestyle='-')
ax.plot(x ,y, marker='*', 
        color='red', 
        linestyle = '--', 
        linewidth = '2', 
        markeredgecolor = 'blue', 
        markerfacecolor = 'orange',
        markersize=12)
ax.set_title("line plot", fontdict= font1, loc='left')
ax.set_xlabel("x", fontdict= font2, loc='right')
ax.set_ylabel('y', fontdict= font2, loc='center')
fig.tight_layout()
```
