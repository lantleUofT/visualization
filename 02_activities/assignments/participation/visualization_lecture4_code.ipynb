```python
#import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y2 = np.array([100,180,240,99,220])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7, 3))
ax1.scatter(x1, y1)
ax2.bar(x2, y2)
fig.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols =2, nrows=1, figsize =(7,3))
ax1.scatter(x1,y1)
ax1.annotate('text', xy=(10,60), xytext=(20,60), arrowprops=dict(facecolor = 'black'))
ax2.bar(x2,y2)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4),
                                    layout = 'constrained')
someaxes["ax1"].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1, y1)
someaxes['ax1'].set_xlabel('text', fontsize = 10)
someaxes['ax2'].set_xlabel('text_y', fontsize = 10)
someaxes['ax3'].set_xlabel('text_ax3', fontsize = 18)
plt.show
```

```python
x = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize=(7, 3))
ax.bar(x, y1, color = 'indigo')
ax.plot(x, y2, color = 'red')
```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color ='red')
ax.errorbar(x, y2, yerr = y2_sd, 
            fmt = 'o-', 
            ecolor = 'indigo', 
            elinewidth= 4, 
            capsize = 6, 
            capthick=4,
            errorevery= 2)
```

```python
from PIL import Image
import requests
from io import BytesIO
```

```python
url = "https://picsum.photos/400/300"

#add headers to website doesn't forbid the action
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "image/*,*/*;q=0.8",
    "Referer": "https://picsum.photos/"
}

response = requests.get(url, headers=headers)
image_file = BytesIO(response.content)
image = Image.open(image_file)
image
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = 'red')
ax_image = fig.add_axes([0.1, 
                         0.11, 
                         0.15, 
                         0.35]) #values are x coord, y coord, width, height in that order
ax_image.imshow(image)
ax_image.axis('off')

path = '/Users/levi/Documents/Documents/Graduate School Stuff/DSI certification class/DSI coding/visualization/02_activities/assignments/participation'
filename = '/image.png'
plt.savefig(path+filename, dpi=300) 
```
