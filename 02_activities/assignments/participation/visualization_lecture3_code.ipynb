```python
#import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#randomly generate data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
#my version
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, y2)
```

```python
#her version
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc = "upper left", frameon = True, fontsize = 12, ncol = 2, shadow = True,
          bbox_to_anchor = (1, 1))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.text(10, 95, "this value is important", ha = 'right', color = "red", size = 20)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.text(10, 95, ". Data:(1, 5)", transform = ax.transData)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform = ax.transAxes)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform = fig.transFigure)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.axis([0, 10, 0, 10])
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform = fig.transFigure)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform = ax.transAxes)
ax.text(1, 5, ". Data:(1, 5)", transform = ax.transData)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.annotate('this is important', xy=(10,95), xytext=(20,94), arrowprops=dict(arrowstyle = 'wedge', color='black'))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
font1 = {'family': 'serif', 'color': 'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")
plt.xticks(rotation=45,ha='right')
plt.xlabel('label', fontsize= 18, fontdict=font1)
```
