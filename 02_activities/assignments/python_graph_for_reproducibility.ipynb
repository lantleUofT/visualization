```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
import seaborn as sns
import matplotlib.ticker as mticker
```

```python
#link to download data: https://data.ontario.ca/dataset/greenhouse-gas-emissions-reporting-by-facility/resource/0996bfd9-ed27-4f78-8ed1-9e024185f10a
pd1 = pd.read_csv('/Users/levi/Documents/Documents/Graduate School Stuff/DSI certification class/DSI coding/visualization/02_activities/assignments/participation/0996bfd9-ed27-4f78-8ed1-9e024185f10a.csv')
pd1.info()
```

```python
df = pd1.groupby(['Year'])['Carbon dioxide (CO2) from non-biomass in CO2e (t)'].sum().reset_index()
df
```

```python
font1 = {'family': 'sans-serif', 'color': 'Green', 'size': '15'}
font2 = {'family': 'sans-serif', 'color': 'Green', 'size': '10'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(df['Year'], df['Carbon dioxide (CO2) from non-biomass in CO2e (t)'], color = 'Green')
ax.set_title('Change in CO2 Emissions Over Time', fontdict=font1)
ax.set_xlabel('Year', fontdict=font2)
ax.set_ylabel('Tonnes of CO2 Emissions (non-biomass)', fontdict=font2)
ax.ticklabel_format(style='plain', axis='y')
ax.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f'{x/1e6:.0f}')
)
ax.set_ylabel('CO₂e (Million Tonnes)')
avg = df['Carbon dioxide (CO2) from non-biomass in CO2e (t)'].mean()
ax.axhline(avg, color='red', linestyle='--', linewidth=1,
           label=f'Average CO2 Emissions')
ax.legend(fontsize=8)
ax.text(0.5, -0.3, "We don't even have data for 2024-2025 yet!",
        ha='center', va='top', color='red', size=10,
        transform=ax.transAxes)
fig.tight_layout()


```
